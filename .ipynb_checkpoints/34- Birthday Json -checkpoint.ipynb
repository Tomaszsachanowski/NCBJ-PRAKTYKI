```python
import json

info_about_me = {
    "name": "Michele",
    "has_a_dog": False
}

with open("info.json", "w") as f:
    json.dump(info_about_me, f)
```

```python
with open("info.json", "r") as f:
    info = json.load(f)

if info["has_a_dog"]:
    print("{} has a dog".format(info["name"]))
else:
    print("{} does not have a dog".format(info["name"]))
```

```python
import json
import calendar
with open("JSON.json", "r") as file:
    dictionary = json.load(file)

file = open("JSON.json", "w")
slowa_klucz = ["tak","nie"]
while True:
    print("Witam w urodziwnowej ksiazce o to nasi ludize")
    for x,y in dictionary.items():
        print(x,y)
    czlowiek = input("Podaj imie i nazwisko: ")
    if czlowiek in dictionary:
        print(czlowiek,dictionary[czlowiek])
    else:
        print("brak takiego czlowieka")
        while True:
            while True:
                try:
                    miesiac = int(input("Podaj mesiac urodzenia"))
                    if miesiac > 0 and miesiac < 13:
                        break
                    else:
                        print("Zla wartosc")
                except ValueError:
                    print("Podaj liczbe")
            while True:
                try:
                    dzien = int(input("Podaj dzien urodzenia"))
                    if dzien > 0 and dzien < 31:
                        break
                    else:
                        print("Zla wartosc")
                except ValueError:
                    print("Podaj liczbe")
            while True:
                try:
                    rok = int(input("Podaj rok urodzenia"))
                    if rok <= 2018:
                        break
                    else:
                        print("Zla wartosc")
                except ValueError:
                    print("Podaj liczbe")
            try:
                calendar.weekday (rok, miesiac,dzien)
                break
            except ValueError:
                print("taki dzien nie istnieje!!! ")
                
        data = [str(dzien),str(miesiac),str(rok)]
        dictionary[czlowiek]="/".join(data)  
    while True:
        dalej = input("Bawisz sie dalej: tak. nie? ")
        if dalej in slowa_klucz:
            break
        else:
            print("Zly wyraz ")
    if dalej == "nie":
        json.dump(dictionary,file)
        file.close()
        break
    else:    
        print(dictionary)
```
